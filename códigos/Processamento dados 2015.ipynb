```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import sqlite3
import json
import os
from pandas.io.json import json_normalize
pd.set_option('display.max_columns', 50)
```

```python
path = "D:\\Dropbox\\Projeto Brasileirao\\caRtola\\data\\2015\\raw\\CartolaFC 2015 - Rodada 38 - 2015-12-11\\38\\"
files = os.listdir(path)
```

```python
data = []
for file in files:
    f = path+file    
    with open(f) as data_file:    
        data.append(json.load(data_file))
```

```python
for x in data[0]:
    print(x)
```

# Processamento 

```python
def prepara_scouts(loaded_json):
    aux = {}
    for i in loaded_json['atleta']:
        for j in i['scout']:
            aux.update({j['nome']:j['quantidade']})
        i['scout'] = aux
        aux = {}
    return loaded_json
```

```python
for load in data:
    prepara_scouts(load)
```

```python
rodada_38 = pd.DataFrame()
for file in data:
    rodada_38 = rodada_38.append(json_normalize(file['atleta']), ignore_index=True)
```

```python
rodada_38.shape
```

```python
rodada_38.duplicated().value_counts()
```

```python
rodada_38.drop_duplicates(inplace=True)
```

```python
rodada_38.columns
```

```python
rodada_38.drop(['apelido', 'clube.abreviacao', 'clube.mercado',
       'clube.nome', 'clube.slug', 'foto',
       'partida_clube_casa.abreviacao', 'partida_clube_casa.id',
       'partida_clube_casa.mercado', 'partida_clube_casa.nome',
       'partida_clube_casa.slug', 'partida_clube_visitante.abreviacao',
       'partida_clube_visitante.id', 'partida_clube_visitante.mercado',
       'partida_clube_visitante.nome', 'partida_clube_visitante.slug',
       'partida_data',  'posicao.abreviacao', 'posicao.id',
       'posicao.nome', 'status', 'status_id'], axis=1, inplace=True)
#rodada_38.drop_duplicates(inplace=True)
```

```python
rodada_38.drop_duplicates(inplace=True)
```

```python
rodada_38['Rodada']=38
```

```python
rodada_38.columns
```

```python
rodada_38.rename(columns={'clube.id': 'ClubeID','id': 'AtletaID', 'jogos':'Jogos',
                                  'pontos': 'Pontos', 'media':'PontosMedia','preco': 'Preco',
                                  'variacao': 'PrecoVariacao','scout.FS': 'FS','scout.PE': 'PE',
                                  'scout.A': 'A', 'scout.FT': 'FT','scout.FD': 'FD', 'scout.FF': 'FF', 
                                  'scout.G': 'G','scout.I': 'I', 'scout.PP':'PP', 'scout.RB': 'RB',
                                  'scout.FC':'FC','scout.GC': 'GC', 'scout.CA':'CA', 'scout.CV':'CV',
                                  'scout.SG': 'SG', 'scout.DD':'DD','scout.DP': 'DP','scout.GS':  'GS'},inplace=True)
```

```python
rodada_38.head()
```

```python
rodada_38 = rodada_38.apply(pd.to_numeric)
```

```python
cols_scouts_def = ['CA','CV','DD','DP','FC','GC','GS','RB','SG'] # alphabetical order
cols_scouts_atk = ['A','FD','FF','FS','FT','G','I','PE','PP'] # alphabetical order
cols_scouts = cols_scouts_def + cols_scouts_atk
```

```python
print('Colunas:', '\n', rodada_38.columns, '\n')
# this data have all player on Cartola dabase 2017, also have ClubeID, nickname and Position
# We will explore
print('Base: \n',rodada_38.head(), '\n')
print('Tamanaho da base: \n ', rodada_38.shape, '\n')
print('Quantidade de duplicados: \n', rodada_38.duplicated().value_counts(),  '\n') # This dataset doesn't have any duplicated
print('Quantidade de Atletas Missing: \n', rodada_38.AtletaID.isnull().sum(), '\n') # We have six positions of players on field , GK, DF, (RB and LF), MF, St, Coach
print('Quantidade de clubes Missing: \n', rodada_38.ClubeID.isnull().sum(),  '\n') # twenty teams
print('Quantidade de jogadores por rodada: ', rodada_38.Rodada.value_counts().sort_index())
print("RODADA 0 É ANTES DO GAME COMEÇAR, JOGADORES COM PONTOS ZERADO E VALOR INICIAL")
rodada_38['Scouts'] = rodada_38[cols_scouts].sum(axis=1)
rodada_38.sort_values(by=['AtletaID','Rodada', 'Jogos', 'Scouts', 'Pontos'],inplace=True)
```

```python
print('Tamanho da base de 2018 antes da limpeza', rodada_38.shape)
rodada_38['Ano'] = 2015
# Eliminando jogadores com nan no nome do clube e que não participaram de jogos
rodada_38 = rodada_38[rodada_38.Scouts>0]
rodada_38 = rodada_38[(rodada_38.PrecoVariacao != 0) | (rodada_38.Pontos!=0)]
rodada_38.drop_duplicates(subset=['AtletaID','ClubeID','Rodada'],keep='last', inplace=True)
print('Tamanho da base com apenas jogadores que jogaram:',rodada_38.shape)
rodada_38_limpo = rodada_38[(~rodada_38.ClubeID.isnull()) & (rodada_38.Rodada>0)]
print('Tamanho da base limpa:', rodada_38_limpo.shape)

rodada_38_limpo.sample(10)
```
