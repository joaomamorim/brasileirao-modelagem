# Processamento 

```python
scouts_2017 = pd.DataFrame()
for i in data.API_ATLETAS_MERCADO:
    loaded_json = json.loads(i)
    scouts_2017 = scouts_2017.append(json_normalize(loaded_json['atletas']), ignore_index=True)
    #loaded_json = 
scouts_2017.drop_duplicates(inplace=True)
```

```python
scouts_2017.drop(columns=['apelido','foto', 'nome', 'scout', 'posicao_id', 'status_id'], inplace=True)
scouts_2017.rename(columns={'rodada_id':'Rodada','jogos_num': 'Jogos', 'clube_id': 'ClubeID','atleta_id': 'AtletaID',
                                  'pontos_num': 'Pontos', 'media_num':'PontosMedia','preco_num': 'Preco',
                                  'variacao_num': 'PrecoVariacao','scout.FS': 'FS','scout.PE': 'PE',
                                  'scout.A': 'A', 'scout.FT': 'FT','scout.FD': 'FD', 'scout.FF': 'FF', 
                                  'scout.G': 'G','scout.I': 'I', 'scout.PP':'PP', 'scout.RB': 'RB',
                                  'scout.FC':'FC','scout.GC': 'GC', 'scout.CA':'CA', 'scout.CV':'CV',
                                  'scout.SG': 'SG', 'scout.DD':'DD','scout.DP': 'DP','scout.GS':  'GS'},inplace=True)
```

```python
cols_scouts_def = ['CA','CV','DD','DP','FC','GC','GS','RB','SG'] # alphabetical order
cols_scouts_atk = ['A','FD','FF','FS','FT','G','I','PE','PP'] # alphabetical order
cols_scouts = cols_scouts_def + cols_scouts_atk
```

```python
scouts_2017.Rodada.value_counts().sort_index()
```

```python
print('Colunas:', '\n', scouts_2017.columns, '\n')
# this data have all player on Cartola dabase 2017, also have ClubeID, nickname and Position
# We will explore
print('Base: \n',scouts_2017.head(), '\n')
print('Tamanaho da base: \n ', scouts_2017.shape, '\n')
print('Quantidade de duplicados: \n', scouts_2017.duplicated().value_counts(),  '\n') # This dataset doesn't have any duplicated
print('Quantidade de Atletas Missing: \n', scouts_2017.AtletaID.isnull().sum(), '\n') # We have six positions of players on field , GK, DF, (RB and LF), MF, St, Coach
print('Quantidade de clubes Missing: \n', scouts_2017.ClubeID.isnull().sum(),  '\n') # twenty teams
scouts_2017['Scouts'] = scouts_2017[cols_scouts].sum(axis=1)
scouts_2017.sort_values(by=['AtletaID','Rodada','Jogos', 'Scouts', 'Pontos'],inplace=True)
scouts_2017.drop_duplicates(subset=['AtletaID','Rodada'],keep='last', inplace=True)
print('Quantidade de jogadores por rodada: ', scouts_2017.Rodada.value_counts().sort_index())
```
