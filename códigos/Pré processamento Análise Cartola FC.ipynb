# Os dados de 2015 estão acumulando, vamos tratar isso

```python
def get_scouts_for_round(df, round_):
    suffixes = ('_curr', '_prev')
    cols_current = [col + suffixes[0] for col in cols_scouts]
    cols_prev = [col + suffixes[1] for col in cols_scouts]
    
    df_round = df[df['Rodada'] == round_]
    if round_ == 1: return df_round
    
    df_round_prev = df[df['Rodada'] < round_].groupby('AtletaID', as_index=False)[cols_scouts].max()
    df_players = df_round.merge(df_round_prev, how='left', on=['AtletaID'], suffixes=suffixes)
    
    # if is the first round of a player, the scouts of previous rounds will be NaNs. Thus, set them to zero
    df_players.fillna(value=0, inplace=True)
    
    # compute the scouts 
    df_players[cols_current] = df_players[cols_current].values - df_players[cols_prev].values
    
    # update the columns
    df_players.drop(labels=cols_prev, axis=1, inplace=True)
    df_players = df_players.rename(columns=dict(zip(cols_current, cols_scouts)))
    df_players.SG = df_players.SG.clip_lower(0)
    
    return df_players
```

```python
n_rounds = scouts_2015_limpo['Rodada'].max()
df_scouts = pd.DataFrame()
if np.isnan(n_rounds):
    scouts_2015_limpo = scouts_2015_limpo
else:
    for i in range(1, n_rounds+1):
        df_round = get_scouts_for_round(scouts_2015_limpo, i)
        print("Dimensões da rodada #{0}: {1}".format(i, df_round.shape))
        df_scouts = df_scouts.append(df_round, ignore_index=True)
    
print(df_scouts.shape)
scouts_2015_limpo = df_scouts.copy()
del(df_scouts)
scouts_2015_limpo.head()
```

### Mudar o padrão dos nomes dos clubes:

    - O estado só será mantido nos clubes que se chamam Atlético

```python
def clubenome(x):
    x = x.replace(' - ', '-')
    a = x.split('-')
    if (a[0]=='Atlético')| (a[0]=='América'):
        return a[0]+'-'+ a[1]
    elif a[0]=='Vasco':
        return a[0]
        
    return a[0]
```

#### Tem clubes com mais gols que finalizações, vamos tratar o número min de finalizações certas sendo o n de gols

```python
def ajeitafinac(gols,finc):
    n = len (gols)
    result = []
    for i in range(0,n):
        if (finc.iloc[i]<gols.iloc[i]):
            result.append(gols.iloc[i])
        else:
            result.append(finc.iloc[i])
    return result
```

```python
result_h = ajeitafinac(bra.placar_casa, bra.FinC_h)
result_a = ajeitafinac(bra.placar_visitante, bra.FinC_a)
bra['FinC_h'] = result_h
bra['FinC_a'] = result_a
bra['Finalizacoes_h'] = bra.FinC_h + bra.FF_h
bra['Finalizacoes_a'] = bra.FinC_a + bra.FF_a
bra['Proporcao_C_h'] = bra.placar_casa/bra.FinC_h
bra['Proporcao_Tot_h'] = bra.placar_casa/bra.Finalizacoes_h
bra['Proporcao_C_a'] = bra.placar_visitante/bra.FinC_a
bra['Proporcao_Tot_a'] = bra.placar_visitante/bra.Finalizacoes_a
bra.fillna(0,inplace=True)
```
