### Medidas de qualidade

```python
def distancia(array1, array2):
    """
    Input: Dois vetores
    Output: Distância entre dois vetores ao quadrado
    """
    tamanho = len(array1)
    dist = 0
    for i in range(tamanho):
        dist += (array1[i] - array2[i]) ** 2
    return dist


def acerto(array1, array2):
    """
    Verifica se o índice do argmax de cada vetor é igual:
    :param array1: vetor
    :param array2: vetor
    :return: 1 ou 0
    """
    return int(array1.argmax() == array2.argmax())


def RPS(array1, array2):
    a1 = np.cumsum(array1)
    a2 = np.cumsum(array2)
    return distancia(a1, a2)/2

def acuracia(ndarray1, ndarray2, rodadas_prevista):
    """   :param ndarray1:  Array com as probabilidade = [[0.52, 0.2, 0.28], [0.32, 0.29, 0.29], ..., [0.3, 0.29, 0.41] ]
          :param ndarray2: Array com os resultados da partida = [[1, 0, 0], [0, 1, 0], ...,[0, 0, 1] ]
          :param rodadas_prevista: Número de rodadas que foram previstas
          [1, 0, 0] = Vitória mandante
          [0, 1, 0] = Empate
          [0, 0, 1] = Vitória visitante
          Saída: Medidade De Finetti, proporção e taxa de sucesso
      """
    jogos = len(ndarray1)  # Número de linhas (partidas)
    definetti, prop, sucesso, rps = 0, 0, 0, 0
    saida: Dict[str, List[Any]] = {'De Finetti': [], 'Proporção': [], 'Proporção acum por rodada': [], \
                                   'Taxa de sucesso':[], 'RPS': []}
    defi_acum = 0
    prop_acum = 0
    sucesso_acum =  0
    rps_acum = 0
    for j in range(jogos):
        definetti += distancia(ndarray1.iloc[j], ndarray2.iloc[j])
        prop += acerto(ndarray1.iloc[j], ndarray2.iloc[j])
        sucesso += np.dot(ndarray1.iloc[j], ndarray2.iloc[j])
        rps += RPS(ndarray1.iloc[j], ndarray2.iloc[j])
        if (j + 1) % 10 == 0:
            defi_acum += definetti
            prop_acum += prop
            sucesso_acum += sucesso
            rps_acum += rps
            saida['De Finetti'].append(definetti/10)
            saida['Proporção'].append(prop/10)
            saida['Proporção acum por rodada'].append(prop_acum/j)
            saida['Taxa de sucesso'].append(sucesso/10)
            saida['RPS'].append(rps/10)
            definetti, prop, sucesso, rps = 0, 0, 0, 0
    saida['De Finetti'].append(defi_acum/(10*rodadas_prevista))
    saida['Proporção'].append(prop_acum/(10*rodadas_prevista))
    saida['Taxa de sucesso'].append(sucesso_acum/(10*rodadas_prevista))
    saida['RPS'].append(rps_acum/(10*rodadas_prevista))
    return saida

def acuracia_col(x,y):
    defi = []
    rps = []
    acer = []
    jogos = len(x)
    
    for jogo in range(jogos):
        defi.append(distancia(x.iloc[jogo],y.iloc[jogo]))
        rps.append(RPS(x.iloc[jogo],y.iloc[jogo]))
        acer.append(acerto(x.iloc[jogo],y.iloc[jogo]))
    return (defi,rps,acer)   
    
```

# Estimação

### Vamos acrescerntar a função peso na verossimilhança

###### Função Tau

$\tau_{\lambda, \mu}(x,y) = =\left\{
                \begin{array}{l}
                  1 - \lambda \mu \rho  &\text{se} \quad x = y = 0 \\
                   1 + \lambda \rho  &\text{se} \quad x = 0, y = 1 \\
                  1 + \lambda \mu   & \text{se} \quad x = 1 , y = 0 \\
                  1 -  \rho  & \text{se} \quad x = y = 1 \\
                  1  & \text{se} \quad \text{c.c}
                \end{array}
              \right. $
              
   \begin{equation*}
   \lambda = \exp(\alpha_i\beta_j \gamma) \\
   \mu = \exp(\alpha_j\beta_i) \\
   max(-1/ \lambda, -1/\mu)  \leq \rho \leq min(1/\lambda \mu, 1)
   \end{equation*}
 
   
   
##### Verossimilhança


\begin{align*}
 L(\mathbf{\theta};(x,y))= ( \tau_{\lambda, \mu}(x,y) \dfrac{\lambda^x exp(-\lambda)}{x!}\dfrac{\mu^y exp(-\mu)}{y!})^{\dfrac{rodada_da_partida}{ultima_rodada}^\frac{1}{2}}
 \end{align*}


#### Função Tau

```python
def rho_correction(x, y, lambda_x, mu_y, rho):
    if x==0 and y==0:
        return 1- (lambda_x * mu_y * rho)
    elif x==0 and y==1:
        return 1 + (lambda_x * rho)
    elif x==1 and y==0:
        return 1 + (mu_y * rho)
    elif x==1 and y==1:
        return 1 - rho
    else:
        return 1.0
```

### Log verossimilhança

```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html

def solve_parameters_DC(dataset, debug = False, init_vals=None, options={'disp': False, 'maxiter':10000000},
                     restricao = [{'type':'eq', 'fun': lambda x: (sum(x[0:20]) -1)}, 
                                 {'type':'eq', 'fun': lambda x: (sum(x[20:40]) -1)}] , ultima_rodada = 19, weight = 2, **kwargs):
    times = np.sort(dataset['home_team'].unique())
    n_times = len(times)
    if init_vals is None:
        # random initialisation of model parameters
        init_vals = np.concatenate((np.random.uniform(0,1,(n_times)), # parâmetros de ataque
                                      np.random.uniform(0,-1,(n_times)), # parâmetros de defesa
                                      np.array([0, 1.0]) # rho (score correction), gamma (home advantage)
                                     ))
    def dc_log_like(x, y, alpha_x, beta_x, alpha_y, beta_y, rho, gamma, rodada, rodada_atual = 19, fator = 2):
        lambda_x, mu_y = exp(alpha_x + beta_y + gamma), exp(alpha_y + beta_x) 
        return (((rodada/rodada_atual)**(1/fator)) * ((np.log(rho_correction(x, y, lambda_x, mu_y, rho)) + 
                np.log(poisson.pmf(x, lambda_x)) + np.log(poisson.pmf(y, mu_y)))))

    def dc_estimativa_parametros(params):
        ataque_param = dict(zip(times, params[:n_times]))
        defesa_param = dict(zip(times, params[n_times:(2*n_times)]))
        rho, gamma = params[-2:]
        log_like = [dc_log_like(row.placar_casa, row.placar_visitante, ataque_param[row.home_team], defesa_param[row.home_team],
                     ataque_param[row.away_team], defesa_param[row.away_team], rho, gamma, row.Rodada, rodada_atual = ultima_rodada,
                               fator = weight) for row in dataset.itertuples()]
        return -sum(log_like)
    opt_output = minimize(dc_estimativa_parametros, init_vals, options=options, constraints = restricao, **kwargs)
    if debug:
        # sort of hacky way to investigate the output of the optimisation process
        return opt_output
    else:
        return -opt_output.fun ,dict(zip(["ataque_"+time for time in times] + 
                        ["defesa_"+time for time in times] +
                        ['rho', 'home_adv'],
                        opt_output.x))
```

##### DIstribuição Poisson Bivariada

A distribuição bivariada de Holgate:

$
P(X=x, Y=y) = e^{-\left(\lambda_1+\lambda_2+\lambda_3\right)} \sum_{i=0}^{min(x,y)} \dfrac{\lambda_1^{x-i}}{(x-i)!}\dfrac{\lambda_2^{y-i}}{(y-i)!}\dfrac{\lambda_3^{i}}{(i)!}
$

### Poisson Bivariada

```python
np.random.seed(14)
x1 = np.random.poisson(1.48,190)
np.random.seed(14)
x2 = np.random.poisson(0.98,190)
np.random.seed(14)
x3 = np.random.poisson(0.01,190)
x = x1 + x3
y = x2 + x3
```

\begin{equation} 
\text{SumF} = \sum_{i=0}^{min(x,y)} \dfrac{\lambda_1^{x-i}}{(x-i)!}\dfrac{\lambda_2^{y-i}}{(y-i)!}\dfrac{\lambda_3^{i}}{(i)!}
\end{equation}

```python
def sumF(hgoals, agoals, lambda_1, lambda_2, lambda_3):
    max_v = min(hgoals,agoals)
    somatorio = 0
    for k in range(0,max_v+1):
        somatorio += (((lambda_1**(hgoals-k))/factorial((hgoals-k))) *
        ((lambda_2**(agoals-k))/factorial((agoals-k))) *
        ((lambda_3**k)/factorial(k)))
    return somatorio
```

```python
def bp_log_lik(xx, yy, lambda_1, lambda_2, lambda_3):
    return ((-lambda_1 - lambda_2 - lambda_3) + np.log(sumF(xx, yy, lambda_1, lambda_2, lambda_3)))
```

```python
def otimizar(param):
    lambda_1 = param[0]
    lambda_2 = param[1]
    lambda_3 = param[2]
    resultado = 0
    for i in range(len(x)):
        resultado += bp_log_lik(x[i],y[i],lambda_1,lambda_2,lambda_3)
    return -resultado 
```

# Poisson Bivariada

```python
def solve_parameters_PB(dataset, debug = False, init_vals=None, options={'disp': False, 'maxiter':100000},
                     restricao = [{'type':'eq', 'fun': lambda x: x[0] -1}, 
                                 {'type':'eq', 'fun': lambda x: x[20] -1}]  , ultima_rodada = 19, weight = 2, **kwargs):
    times = np.sort(dataset['home_team'].unique())
    n_times = len(times)
    if init_vals is None:
        # parametros iniciais
        init_vals = np.concatenate((np.random.uniform(0,1,(n_times)), # attack strength
                                      np.random.uniform(0,-1,(n_times)), # defence strength
                                      np.array([0.05, 0.64]) # lambda_3 (score correction), home (home advantage)
                                     ))
    def sumF(hgoals, agoals, lambda_1, lambda_2, lambda_3):
        max_v = min(hgoals,agoals)
        somatorio = 0
        for k in range(0,max_v+1):
            somatorio += (((lambda_1**(hgoals-k))/factorial((hgoals-k))) *
            ((lambda_2**(agoals-k))/factorial((agoals-k))) *
            ((lambda_3**k)/factorial(k)))
        return somatorio
    
    def bp_log_lik(xx, yy, alpha_x, beta_x, alpha_y, beta_y, gamma, home, rodada, rodada_atual=19, fator= 2):
        lambda_1, lambda_2, lambda_3 = exp(alpha_x + beta_y + home), exp(alpha_y + beta_x), exp(gamma)
        return  ((rodada/rodada_atual)**(1/fator)) * ((-lambda_1 - lambda_2 - lambda_3) + np.log(sumF(xx, yy, lambda_1, lambda_2, lambda_3)))
    
    def bp_estimativa_parametros(params):
        ataque_param = dict(zip(times, params[:n_times]))
        defesa_param = dict(zip(times, params[n_times:(2*n_times)]))
        lambda_3, home = params[-2:]
        log_like = [bp_log_lik(row.placar_casa, row.placar_visitante, ataque_param[row.home_team], defesa_param[row.home_team],
                     ataque_param[row.away_team], defesa_param[row.away_team], lambda_3, home, row.Rodada, rodada_atual = ultima_rodada,
                               fator = weight) for row in dataset.itertuples()]
        return -sum(log_like)
    opt_output = minimize(bp_estimativa_parametros, init_vals, options=options, constraints = restricao , method = 'L-BFGS-B', tol= 1e-8)
    if debug:
        # sort of hacky way to investigate the output of the optimisation process
        return opt_output
    else:
        return -opt_output.fun, dict(zip(["ataque_"+time for time in times] + 
                        ["defesa_"+time for time in times] +
                        ['rho', 'home_adv'],
                        opt_output.x))
```

### Poisson Independente


\begin{align*}
 P(X=x, Y=y) = (  \dfrac{\lambda^x exp(-\lambda)}{x!}\dfrac{\mu^y exp(-\mu)}{y!})^{\dfrac{rodada_da_partida}{ultima_rodada}^\frac{1}{2}}
 \end{align*}

```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html

def solve_parameters_PD(dataset, debug = False, init_vals=None, options={'disp': False, 'maxiter':100000},
                     restricao = [{'type':'eq', 'fun': lambda x: (sum(x[0:20]) -1)}, 
                                 {'type':'eq', 'fun': lambda x: (sum(x[20:40]) -1)}]  , ultima_rodada = 19, weight = 2, **kwargs):
    times = np.sort(dataset['home_team'].unique())
    n_times = len(times)
    if init_vals is None:
        # parametros iniciais
        init_vals = np.concatenate((np.random.uniform(0,1,(n_times)), # Ataque
                                      np.random.uniform(0,-1,(n_times)), # Defesa
                                      np.array([0.64]) #  home (home advantage)
                                     ))
    def log_verossi(x,y,alpha_x, beta_x, alpha_y, beta_y, home, rodada, rodada_atual= 19, fator= 2):
        lambda1 = exp(alpha_x + beta_y + home)
        lambda2 = exp(alpha_y + beta_x)
        return ((rodada/rodada_atual)**(1/weight)) * (np.log(poisson.pmf(x, lambda1)) + np.log(poisson.pmf(y, lambda2)))
    def pd_estimativa_parametros(params):
        ataque_param = dict(zip(times, params[:n_times]))
        defesa_param = dict(zip(times, params[n_times:(2*n_times)]))
        home = params[-1]
        log_like = [log_verossi(row.placar_casa, row.placar_visitante, ataque_param[row.home_team], defesa_param[row.home_team],
                     ataque_param[row.away_team], defesa_param[row.away_team], home, row.Rodada, rodada_atual = ultima_rodada,
                               fator = weight) for row in dataset.itertuples()]
        return -sum(log_like)
    opt_output = minimize(pd_estimativa_parametros, init_vals, options=options, constraints = restricao)
    if debug:
        # sort of hacky way to investigate the output of the optimisation process
        return opt_output
    else:
        return opt_output.fun, dict(zip(["ataque_"+time for time in times] + 
                        ["defesa_"+time for time in times] +
                        ['home_adv'],
                        opt_output.x))
```

```python
def retorna_param_aut(time_casa, time_visi, param_PB,  param_DC,
                      gol_m, gol_FC_m, gol_FC_RB_m, gol_F_m, gol_F_RB_m,
                      finc_m, logit_m, scouts, prop):
    parametros = {}
    
    
    lamb1 = finc_m.predict(pd.DataFrame(data={'team': str(time_casa), 'opponent': str(time_visi),
                                       'home':1},index=[1]))
    lamb2 = finc_m.predict(pd.DataFrame(data={'team': str(time_visi), 'opponent': str(time_casa),
                                       'home':0},index=[1]))
    p1 = logit_m.predict(pd.DataFrame(data={'team': str(time_casa), 'opponent': str(time_visi),
                                       'home':1},index=[1]))
    p2 = logit_m.predict(pd.DataFrame(data={'team': str(time_visi), 'opponent': str(time_casa),
                                       'home':0},index=[1]))
    p11 = prop.loc[str(time_casa),'prop_C']
    
    p22 = prop.loc[str(time_visi),'prop_C']
    
     
    parametros.update({'FinC': [lamb1, lamb2, p1, p2, p11, p22]})
    #---
    lambda_1_PB = exp(param_PB['ataque_'+time_casa] + param_PB['defesa_'+ time_visi] + param_PB['home_adv'])
    lambda_2_PB = exp(param_PB['ataque_'+time_visi] + param_PB['defesa_'+ time_casa] )
    lambda_3_PB = exp(param_PB['rho'])
        
    parametros.update({'PB':[lambda_1_PB, lambda_2_PB, lambda_3_PB]}) 
    #---
     
    lambda_1_PD_gols = gol_m.predict(pd.DataFrame(data={'team': str(time_casa), 'opponent': str(time_visi),
                                                                    'home':1},index=[1]))
    lambda_2_PD_gols = gol_m.predict(pd.DataFrame(data={'team': str(time_visi), 'opponent': str(time_casa),
                                                                    'home':0},index=[1]))
    
    parametros.update({'PD_gols': [lambda_1_PD_gols,lambda_2_PD_gols]})
    #---
    
    lambda_1_PD_gols_FinC = gol_FC_m.predict(pd.DataFrame(data={'team': str(time_casa), 'opponent': str(time_visi),
                                                                    'FinC':scouts.loc[str(time_casa),'FinC'],'home':1},index=[1]))
    lambda_2_PD_gols_FinC = gol_FC_m.predict(pd.DataFrame(data={'team': str(time_visi), 'opponent': str(time_casa),
                                                                    'FinC':scouts.loc[str(time_visi),'FinC'],'home':0},index=[1]))
          
    parametros.update({'PD_gols_FinC': [lambda_1_PD_gols_FinC,lambda_2_PD_gols_FinC]})
    
    #---
    lambda_1_PD_gols_FinC_RB = gol_FC_RB_m.predict(pd.DataFrame(data={'team': str(time_casa), 'opponent': str(time_visi),
                                                                    'FinC':scouts.loc[str(time_casa),'FinC'],'home':1
                                                               ,'RB':scouts.loc[str(time_casa),'RB']},index=[1]))
    lambda_2_PD_gols_FinC_RB = gol_FC_RB_m.predict(pd.DataFrame(data={'team': str(time_visi), 'opponent': str(time_casa),
                                                                    'FinC':scouts.loc[str(time_visi),'FinC'],'home':0
                                                               ,'RB':scouts.loc[str(time_visi),'RB']},index=[1]))
          
    parametros.update({'PD_gols_FinC_RB': [lambda_1_PD_gols_FinC_RB,lambda_2_PD_gols_FinC_RB]})
    
    #---
    lambda_1_PD_gols_Fina = gol_F_m.predict(pd.DataFrame(data={'team': str(time_casa), 'opponent': str(time_visi),
                                                                    'Finalizacoes':scouts.loc[str(time_casa),'Finalizacoes'],'home':1},index=[1]))
    lambda_2_PD_gols_Fina = gol_F_m.predict(pd.DataFrame(data={'team': str(time_visi), 'opponent': str(time_casa),
                                                                    'Finalizacoes':scouts.loc[str(time_visi),'Finalizacoes'],'home':0},index=[1]))
          
    parametros.update({'PD_gols_Fina': [lambda_1_PD_gols_Fina,lambda_2_PD_gols_Fina]})
    
    #---
    lambda_1_PD_gols_Fina_RB = gol_F_RB_m.predict(pd.DataFrame(data={'team': str(time_casa), 'opponent': str(time_visi),
                                                                    'Finalizacoes':scouts.loc[str(time_casa),'Finalizacoes'],
                                                                     'home':1,'RB':scouts.loc[str(time_casa),'RB']},index=[1]))
    lambda_2_PD_gols_Fina_RB = gol_F_RB_m.predict(pd.DataFrame(data={'team': str(time_visi), 'opponent': str(time_casa),
                                                                    'Finalizacoes':scouts.loc[str(time_visi),'Finalizacoes'],
                                                                     'home':0,'RB':scouts.loc[str(time_visi),'RB']},index=[1]))
          
    parametros.update({'PD_gols_Fina_RB': [lambda_1_PD_gols_Fina_RB,lambda_2_PD_gols_Fina_RB]})
    
    #---
    
    lambda_1_DC = exp(param_DC['ataque_'+time_casa] + param_DC['defesa_'+ time_visi] + param_DC['home_adv'])
    lambda_2_DC = exp(param_DC['ataque_'+time_visi] + param_DC['defesa_'+ time_casa] )
    rho = param_DC['rho']
    
    parametros.update({'DC':[lambda_1_DC,lambda_2_DC,rho]})
          
    return parametros
        

def bp_prob(xx, yy, lambda_1, lambda_2, lambda_3):
        #lambda_1, lambda_2, lambda_3 = np.exp(alpha_x + beta_y + home), np.exp(alpha_y + beta_x), np.exp(gamma)
        
        return  ( exp(-lambda_1 - lambda_2 - lambda_3) * (sumF(xx, yy, lambda_1, lambda_2, lambda_3)))
def dc_prob(xx,yy,lambda_1,lambda_2,rho):
    return(rho_correction(xx,yy,lambda_1, lambda_2,rho)*poisson.pmf(xx,lambda_1)*poisson.pmf(yy,lambda_2))
    
def simulate_matches_v(data, param_PB, param_DC, gol_m, gol_FC_m, gol_FC_RB_m, gol_F_m, gol_F_RB_m, finc_m,
                       logit_m, scouts, prop,
                       max_gols=5, max_finc = 16, decimais = 4):
    home = [] # Time casa
    away = [] # Time visitante
    rodada = [] # rodada
    
    
    prob_PD_gol = [] # Modelo 1 
    prob_PD_gol_Fina_RB = [] # Modelo 2
    prob_PD_gol_FinC_RB = [] # Modelo 3
    prob_PD_gol_Fina = [] # Modelo 4
    prob_PD_gol_FinC = [] # Modelo *4*
    prob_binpo_p = [] # Modelo 5
    prob_binpo_log = [] # Modelo *5*
    prob_DC = [] # Modelo 6
    prob_PB = [] # Modelo 7
    
    
    
    
    prob_PD_gol_placar = [] # Modelo 1
    prob_PD_gol_Fina_RB_placar = [] # Modelo 2
    prob_PD_gol_FinC_RB_placar = [] # Modelo 3
    prob_PD_gol_FinC_placar = [] # Modelo 4*
    prob_PD_gol_Fina_placar = [] # Modelo 4*
    prob_binpo_p_placar = [] # Modelo 5
    prob_binpo_log_placar = [] # Modelo *5*
    prob_DC_placar = [] # Modelo 6
    prob_PB_placar = [] # Modelo 7
    

    
    for partida in data.itertuples():
        home.append(partida.home_team)
        away.append(partida.away_team)
        rodada.append(partida.Rodada)
        probx = []
        proby = []
        #--
        params = retorna_param_aut(partida.home_team, partida.away_team,param_PB,param_DC,gol_m,
                                   gol_FC_m, gol_FC_RB_m, gol_F_m, gol_F_RB_m,finc_m, logit_m, scouts, prop)
        #--
        resultado = {}
        parcial = []
        pc, pv = partida.placar_casa, partida.placar_visitante
        
        #---
        
        for i in range(0,max_gols+1):
            for j in range(0, max_gols +1):
                parcial.append(bp_prob(i,j,params['PB'][0], params['PB'][1], params['PB'][2]))
        resultado.update({'PB': np.array(parcial).reshape((max_gols+1, max_gols+1))}) 
        prob_PB_placar.append(round(bp_prob(pc,pv,params['PB'][0], params['PB'][1], params['PB'][2]),4))
        parcial = []
        
       #---
        
        for i in range(0,max_gols+1):
            for j in range(0, max_gols +1):
                parcial.append(poisson.pmf(i,params['PD_gols'][0])*poisson.pmf(j,params['PD_gols'][1]))
        resultado.update({'PD_gols': np.array(parcial).reshape((max_gols+1, max_gols+1))})
        prob_PD_gol_placar.append(float(poisson.pmf(pc,params['PD_gols'][0])*poisson.pmf(pv,params['PD_gols'][1])))
        parcial = []
        
        #---
        
        for i in range(0,max_gols+1):
            for j in range(0, max_gols +1):
                parcial.append(poisson.pmf(i,params['PD_gols_FinC'][0])*poisson.pmf(j,params['PD_gols_FinC'][1]))
        resultado.update({'PD_gols_FinC': np.array(parcial).reshape((max_gols+1, max_gols+1))}) 
        prob_PD_gol_FinC_placar.append(float(poisson.pmf(pc,params['PD_gols_FinC'][0])*poisson.pmf(pv,
                                                                                    params['PD_gols_FinC'][1])))
        parcial = []
        #---
        
        for i in range(0,max_gols+1):
            for j in range(0, max_gols +1):
                parcial.append(poisson.pmf(i,params['PD_gols_FinC_RB'][0])*poisson.pmf(j,params['PD_gols_FinC_RB'][1]))
        resultado.update({'PD_gols_FinC_RB': np.array(parcial).reshape((max_gols+1, max_gols+1))}) 
        prob_PD_gol_FinC_RB_placar.append(float(poisson.pmf(pc,params['PD_gols_FinC_RB'][0])*poisson.pmf(pv,
                                                                                    params['PD_gols_FinC_RB'][1])))
        parcial = []
        #---
        
        for i in range(0,max_gols+1):
            for j in range(0, max_gols +1):
                parcial.append(poisson.pmf(i,params['PD_gols_Fina'][0])*poisson.pmf(j,params['PD_gols_Fina'][1]))
        resultado.update({'PD_gols_Fina': np.array(parcial).reshape((max_gols+1, max_gols+1))}) 
        prob_PD_gol_Fina_placar.append(float(poisson.pmf(pc,params['PD_gols_Fina'][0])*poisson.pmf(
            pv,params['PD_gols_Fina'][1])))
        parcial = []
        #---
        
        for i in range(0,max_gols+1):
            for j in range(0, max_gols +1):
                parcial.append(poisson.pmf(i,params['PD_gols_Fina_RB'][0])*poisson.pmf(j,params['PD_gols_Fina_RB'][1]))
        resultado.update({'PD_gols_Fina_RB': np.array(parcial).reshape((max_gols+1, max_gols+1))}) 
        prob_PD_gol_Fina_RB_placar.append(float(poisson.pmf(pc,params['PD_gols_Fina_RB'][0])*poisson.pmf(
            pv,params['PD_gols_Fina_RB'][1])))
        parcial = []
        #---
        
        for i in range(0,max_gols+1):
            for j in range(0, max_gols +1):
                parcial.append(dc_prob(i,j, params['DC'][0], params['DC'][1], params['DC'][2]))
        resultado.update({'DC': np.array(parcial).reshape((max_gols+1, max_gols+1))}) 
        prob_DC_placar.append(dc_prob(pc,pv, params['DC'][0], params['DC'][1], params['DC'][2]))
        del(parcial)
        #---
        
        # Modelo Bino*Poi
        
        #time casa
        for i in range(max_gols+1):
            s = 0 
            for j in range(max_finc+1):
                s+= binom.pmf(i,j, params['FinC'][2])*poisson.pmf(j,params['FinC'][0])
            probx.append(s)
        
        #time visitante
        for i in range(max_gols+1):
            s = 0 
            for j in range(max_finc+1):
                s+= binom.pmf(i,j, params['FinC'][3])*poisson.pmf(j,params['FinC'][1])
            proby.append(s)
            
        
    
        M = np.outer(np.array(probx), np.array(proby))
        prob_binpo_log_placar.append(M[pc,pv])
        #print(M.shape)
        a = np.tril(M,-1).sum().round(decimais)
        b = np.diag(M).sum().round(decimais)
        c = np.triu(M,1).sum().round(decimais)
        prob_binpo_log.append(np.array([a,b,c]))
        
        probx = []
        proby = []
        # p efeito fixo
        for i in range(max_gols+1):
            s = 0 
            for j in range(max_finc+1):
                s+= binom.pmf(i,j, params['FinC'][4])*poisson.pmf(j,params['FinC'][0])
            probx.append(s)
        
        #time visitante
        for i in range(max_gols+1):
            s = 0 
            for j in range(max_finc+1):
                s+= binom.pmf(i,j, params['FinC'][5])*poisson.pmf(j,params['FinC'][1])
            proby.append(s)
    
        M = np.outer(np.array(probx), np.array(proby))
        prob_binpo_p_placar.append(M[pc,pv])
       
        
        a = np.tril(M,-1).sum().round(decimais)
        b = np.diag(M).sum().round(decimais)
        c = np.triu(M,1).sum().round(decimais)
        prob_binpo_p.append(np.array([a,b,c]))
        
        #--
        a = np.tril(resultado['PB'],-1).sum().round(decimais)
        b = np.diag(resultado['PB']).sum().round(decimais)
        c = np.triu(resultado['PB'],1).sum().round(decimais)
        prob_PB.append(np.array([a,b,c]))
        
        a = np.tril(resultado['PD_gols'],-1).sum().round(decimais)
        b = np.diag(resultado['PD_gols']).sum().round(decimais)
        c = np.triu(resultado['PD_gols'],1).sum().round(decimais)
        prob_PD_gol.append(np.array([a,b,c]))
        
        a = np.tril(resultado['PD_gols_FinC'],-1).sum().round(decimais)
        b = np.diag(resultado['PD_gols_FinC']).sum().round(decimais)
        c = np.triu(resultado['PD_gols_FinC'],1).sum().round(decimais)
        prob_PD_gol_FinC.append(np.array([a,b,c]))
        
        a = np.tril(resultado['PD_gols_FinC_RB'],-1).sum().round(decimais)
        b = np.diag(resultado['PD_gols_FinC_RB']).sum().round(decimais)
        c = np.triu(resultado['PD_gols_FinC_RB'],1).sum().round(decimais)
        prob_PD_gol_FinC_RB.append(np.array([a,b,c]))
        
        a = np.tril(resultado['PD_gols_Fina'],-1).sum().round(decimais)
        b = np.diag(resultado['PD_gols_Fina']).sum().round(decimais)
        c = np.triu(resultado['PD_gols_Fina'],1).sum().round(decimais)
        prob_PD_gol_Fina.append(np.array([a,b,c]))
        
        
        a = np.tril(resultado['PD_gols_Fina_RB'],-1).sum().round(decimais)
        b = np.diag(resultado['PD_gols_Fina_RB']).sum().round(decimais)
        c = np.triu(resultado['PD_gols_Fina_RB'],1).sum().round(decimais)
        prob_PD_gol_Fina_RB.append(np.array([a,b,c]))
        
        a = np.tril(resultado['DC'],-1).sum().round(decimais)
        b = np.diag(resultado['DC']).sum().round(decimais)
        c = np.triu(resultado['DC'],1).sum().round(decimais)
        prob_DC.append(np.array([a,b,c]))
        
    
        
    return pd.DataFrame({'Home':home, 'Away': away, 'Rodada':rodada, 'prob_PB':prob_PB, 
                        'prob_PD_gol':prob_PD_gol, 'prob_PD_gol_FinC_RB':prob_PD_gol_FinC_RB, 'prob_PD_gol_Fina_RB':prob_PD_gol_Fina_RB,
                        'prob_PD_gol_FinC':prob_PD_gol_FinC, 'prob_PD_gol_Fina':prob_PD_gol_Fina,
                        'prob_DC':prob_DC,'prob_binpo_log':prob_binpo_log, 'prob_binpo_p':prob_binpo_p,
                        'prob_PB_placar':prob_PB_placar, 'prob_PD_gol_placar':prob_PD_gol_placar, 'prob_PD_gol_FinC_placar':prob_PD_gol_FinC_placar,
                        'prob_PD_gol_Fina_placar':prob_PD_gol_Fina_placar ,'prob_PD_gol_Fina_RB_placar':prob_PD_gol_Fina_RB_placar, 'prob_DC_placar':prob_DC_placar,
                        'prob_binpo_log_placar':prob_binpo_log_placar, 'prob_binpo_p_placar':prob_binpo_p_placar,'prob_PD_gol_FinC_RB_placar':prob_PD_gol_FinC_RB_placar})[['Home','Away','Rodada', 
                        'prob_PD_gol', 'prob_PD_gol_Fina_RB','prob_PD_gol_FinC_RB','prob_PD_gol_Fina',
                        'prob_PD_gol_FinC','prob_binpo_log','prob_DC','prob_PB','prob_PD_gol_placar',
                        'prob_PD_gol_Fina_RB_placar', 'prob_PD_gol_FinC_RB_placar','prob_PD_gol_Fina_placar',
                        'prob_PD_gol_FinC_placar','prob_binpo_log_placar',
                         'prob_DC_placar', 'prob_PB_placar','prob_binpo_p','prob_binpo_p_placar']] 
           
```

```python
### Estima cada um dos modelo a cada rodada e gera previsões da rodada seguinte

def previssoes_1passo(base, rodada_inicial, rodada_final, estat = False, estat_p = False, **kwards):
    
    if estat ==False:
    
        # Inicializacao
        data_prev, forecast = base[base['Rodada']<=rodada_inicial] , base[base['Rodada'] == rodada_inicial+1]
        data_prev_sta = slice_data_stat(base,rodada_inicial)

        if estat_p:
            scouts = calcula_fin_rb(data_prev_sta,primeiro_turno = True)
            prop = calcula_prop(data_prev_sta, primeiro_turno = True)
        else:
            scouts = calcula_fin_rb(data_prev_sta,primeiro_turno = False)
            prop = calcula_prop(data_prev_sta, primeiro_turno = False)

        hist_param = {'PB': defaultdict(list), 'PD_gol': defaultdict(list),'PD_gol_FinC': defaultdict(list),
                      'PD_gol_Fina': defaultdict(list),'DC': defaultdict(list), 'PD_finC': defaultdict(list),
                      'PD_gol_FinC_rb': defaultdict(list), 'PD_gol_Fina_rb': defaultdict(list),
                      'Prop_logit': defaultdict(list),'Prop': defaultdict(list)}
        previsoes = pd.DataFrame()


        #---
        for time in prop.index:
            hist_param['Prop'][str(time)].append(prop.loc[str(time),'prop_C'])


        #---    
        logit_model_Prop = smf.glm(formula= "Proporcao_C ~ 0 + home + team + opponent", data=data_prev_sta, 
                            family=sm.families.Binomial()).fit()
        prop_parametros = logit_model_Prop.params


        for par in prop_parametros.index:
            hist_param['Prop_logit'][str(par)].append(1)#prop_param[str(par)])


        #---
        poisson_model_FinC = smf.glm(formula= "FinC ~ 0 + home + team + opponent", data=data_prev_sta,
                                     family=sm.families.Poisson()).fit()
        PD_finC = poisson_model_FinC.params

        for par in PD_finC.index:
            hist_param['PD_finC'][str(par)].append(PD_finC[str(par)])   

        #---
        param_PB = solve_parameters_PB(data_prev,ultima_rodada=rodada_inicial,debug=False)[1]


        for par in param_PB.keys():
            hist_param['PB'][str(par)].append(param_PB[str(par)])

        #---    
         
        param_DC = solve_parameters_DC(data_prev,ultima_rodada=rodada_inicial,debug=False)[1]

        for par in param_PB.keys():
            hist_param['DC'][str(par)].append(param_DC[str(par)])


        #---
        poisson_model_gol = smf.glm(formula="goals ~ 0 + home + team + opponent", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
        PD_gol = poisson_model_gol.params

        for par in PD_gol.index:
            hist_param['PD_gol'][str(par)].append(PD_gol[str(par)])


        #---
        poisson_model_gol_finc = smf.glm(formula="goals ~ 0 + home + team + opponent + FinC ", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
        PD_gol_FinC = poisson_model_gol_finc.params

        for par in PD_gol_FinC.index:
            hist_param['PD_gol_FinC'][str(par)].append(PD_gol_FinC[str(par)])

        #---
        poisson_model_gol_finc_rb = smf.glm(formula="goals ~ 0 + home + team + opponent + FinC + RB", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
        PD_gol_FinC_rb = poisson_model_gol_finc_rb.params

        for par in PD_gol_FinC_rb.index:
            hist_param['PD_gol_FinC_rb'][str(par)].append(PD_gol_FinC_rb[str(par)])

        #---
        poisson_model_gol_fina = smf.glm(formula="goals ~ 0 + home + team + opponent + Finalizacoes ", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
        PD_gol_Fina= poisson_model_gol_fina.params

        for par in PD_gol_Fina.index:
            hist_param['PD_gol_Fina'][str(par)].append(PD_gol_Fina[str(par)]) 


        #---
        poisson_model_gol_fina_rb = smf.glm(formula="goals ~ 0 + home + team + opponent + Finalizacoes + RB ", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
        PD_gol_Fina_rb= poisson_model_gol_fina_rb.params

        for par in PD_gol_Fina_rb.index:
            hist_param['PD_gol_Fina_rb'][str(par)].append(PD_gol_Fina_rb[str(par)]) 

        #----------


        previsoes = previsoes.append(simulate_matches_v(forecast, param_PB, param_DC,poisson_model_gol,
                                                        poisson_model_gol_finc,poisson_model_gol_finc_rb,
                                                        poisson_model_gol_fina,poisson_model_gol_fina_rb,
                                                        poisson_model_FinC,logit_model_Prop
                                                        ,scouts, prop,7,20,4),ignore_index=True)
        #-----------------------------------------------------------------------

        for rodada in tqdm(range(rodada_inicial+1, rodada_final)):

            print(rodada)
            data_prev, forecast = base[base['Rodada']<=rodada] , base[base['Rodada']==rodada+1]
            data_prev_sta = slice_data_stat(base,rodada)
            #--

            if estat_p:
                scouts = calcula_fin_rb(data_prev_sta,primeiro_turno = True)
                prop = calcula_prop(data_prev_sta, primeiro_turno = True)
            else:
                scouts = calcula_fin_rb(data_prev_sta,primeiro_turno = False)
                prop = calcula_prop(data_prev_sta, primeiro_turno = False)

            #---
            for time in prop.index:
                hist_param['Prop'][str(time)].append(prop.loc[str(time), 'prop_C'])


            #---    
            logit_model_Prop = smf.glm(formula= "Proporcao_C ~ 0 + home + team + opponent", data=data_prev_sta, 
                            family=sm.families.Binomial()).fit()
            prop_param = logit_model_Prop.params

            for par in prop_param.index:
                hist_param['Prop_logit'][str(par)].append(prop_param[str(par)])



            #---       
            poisson_model_FinC = smf.glm(formula= "FinC ~ 0 + home + team + opponent", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
            PD_finC = poisson_model_FinC.params

            for par in PD_finC.index:
                hist_param['PD_finC'][str(par)].append(PD_finC[str(par)])


            #---
            param_PB = solve_parameters_PB(data_prev,ultima_rodada=rodada,
                                           init_vals=np.fromiter(param_PB.values(), dtype=float), debug=False)[1]

            for par in param_PB.keys():
                hist_param['PB'][str(par)].append(param_PB[str(par)])


            #---    
            param_DC = solve_parameters_DC(data_prev,ultima_rodada=rodada,
                                           init_vals=np.fromiter(param_DC.values(), dtype=float),debug=False)[1]

            for par in param_DC.keys():
                hist_param['DC'][str(par)].append(param_DC[str(par)])


            #---
            poisson_model_gol = smf.glm(formula="goals ~ 0 + home + team + opponent", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
            PD_gol = poisson_model_gol.params

            for par in PD_gol.index:
                hist_param['PD_gol'][str(par)].append(PD_gol[str(par)])


            #---
            poisson_model_gol_finc = smf.glm(formula="goals ~ 0 + home + team + opponent + FinC ", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
            PD_gol_FinC = poisson_model_gol_finc.params

            for par in PD_gol_FinC.index:
                hist_param['PD_gol_FinC'][str(par)].append(PD_gol_FinC[str(par)])


            #---
            poisson_model_gol_finc_rb = smf.glm(formula="goals ~ 0 + home + team + opponent + FinC + RB", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
            PD_gol_FinC_rb = poisson_model_gol_finc_rb.params

            for par in PD_gol_FinC_rb.index:
                hist_param['PD_gol_FinC_rb'][str(par)].append(PD_gol_FinC_rb[str(par)])


            #---


            poisson_model_gol_fina = smf.glm(formula="goals ~ 0 + home + team + opponent + Finalizacoes ", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
            PD_gol_Fina= poisson_model_gol_fina.params

            #---
            for par in PD_gol_Fina.index:
                hist_param['PD_gol_Fina'][str(par)].append(PD_gol_Fina[str(par)]) 

            #-----

            poisson_model_gol_fina_rb = smf.glm(formula="goals ~ 0 + home + team + opponent + Finalizacoes + RB ", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
            PD_gol_Fina_rb= poisson_model_gol_fina_rb.params

            for par in PD_gol_Fina_rb.index:
                hist_param['PD_gol_Fina_rb'][str(par)].append(PD_gol_Fina_rb[str(par)])

            #--------------
            previsoes = previsoes.append(simulate_matches_v(forecast,param_PB, param_DC,poisson_model_gol,
                                                        poisson_model_gol_finc,poisson_model_gol_finc_rb,
                                                        poisson_model_gol_fina,poisson_model_gol_fina_rb,
                                                        poisson_model_FinC,logit_model_Prop
                                                        ,scouts, prop,7,20,4),ignore_index=True)
        
        resultado =  (previsoes, hist_param)
    else:
        # Inicializacao
        data_prev, forecast = slice_data(base,rodada_inicial)
        data_prev_sta = slice_data_stat(base,rodada_inicial)

        if estat_p:
            scouts = calcula_fin_rb(data_prev_sta,primeiro_turno = True)
            prop = calcula_prop(data_prev_sta, primeiro_turno = True)
        else:
            scouts = calcula_fin_rb(data_prev_sta,primeiro_turno = False)
            prop = calcula_prop(data_prev_sta, primeiro_turno = False)

 
        previsoes = pd.DataFrame()
    
        ano = data_prev.Ano.iloc[0]
        path = 'D:\\Dropbox\Projeto Brasileirao\\resultados\\Dissertacao\\acuracia\\'
        file = 'bic_aic_hpassos_' + str(ano) + '_' + str(rodada_inicial) + '.txt'
        
        f = open(path+file,"w+")
        
        #---    
        logit_model_Prop = smf.glm(formula= "Proporcao_C ~ 0 + home + team + opponent", data=data_prev_sta, 
                            family=sm.families.Binomial()).fit()
        prop_parametros = logit_model_Prop.params
        
        

           #---
        poisson_model_FinC = smf.glm(formula= "FinC ~ 0 + home + team + opponent", data=data_prev_sta,
                                     family=sm.families.Poisson()).fit()
        PD_finC = poisson_model_FinC.params
        
        
       

        #---
        Lov_PB, param_PB = solve_parameters_PB(data_prev,ultima_rodada=rodada_inicial,debug=False)
        
        #---   
        Lov_DC, param_DC = solve_parameters_DC(data_prev,ultima_rodada=rodada_inicial,debug=False)
        
        #---
        poisson_model_gol = smf.glm(formula="goals ~ 0 + home + team + opponent", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
        PD_gol = poisson_model_gol.params
       
        #---
        poisson_model_gol_finc = smf.glm(formula="goals ~ 0 + home + team + opponent + FinC ", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
        PD_gol_FinC = poisson_model_gol_finc.params
        
        #---
        poisson_model_gol_finc_rb = smf.glm(formula="goals ~ 0 + home + team + opponent + FinC + RB", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
        PD_gol_FinC_rb = poisson_model_gol_finc_rb.params
        
        #---
        poisson_model_gol_fina = smf.glm(formula="goals ~ 0 + home + team + opponent + Finalizacoes ", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
        PD_gol_Fina= poisson_model_gol_fina.params

        #---
        poisson_model_gol_fina_rb = smf.glm(formula="goals ~ 0 + home + team + opponent + Finalizacoes + RB ", data=data_prev_sta, 
                            family=sm.families.Poisson()).fit()
        PD_gol_Fina_rb= poisson_model_gol_fina_rb.params
        
        #-----------
        f.write("AIC_PD_FINC: {} \n".format(poisson_model_FinC.aic))
        f.write("AIC_PD_GOL: {} \n".format(poisson_model_gol.aic))
        f.write("AIC_PD_GOL_FINC: {} \n".format(poisson_model_gol_finc.aic))
        f.write("AIC_PD_GOL_FINC_RB: {} \n".format(poisson_model_gol_finc_rb.aic))
        f.write("AIC_PD_GOL_FINA: {} \n".format(poisson_model_gol_fina.aic))
        f.write("AIC_PD_GOL_FINA_RB: {} \n".format(poisson_model_gol_fina_rb.aic))
        
        #--
        f.write("BIC_PD_FINC: {} \n".format(poisson_model_FinC.bic))
        f.write("BIC_PD_GOL: {} \n".format(poisson_model_gol.bic))
        f.write("BIC_PD_GOL_FINC: {} \n".format(poisson_model_gol_finc.bic))
        f.write("BIC_PD_GOL_FINC_RB: {} \n".format(poisson_model_gol_finc_rb.bic))
        f.write("BIC_PD_GOL_FINA: {} \n".format(poisson_model_gol_fina.bic))
        f.write("BIC_PD_GOL_FINA_RB: {} \n".format(poisson_model_gol_fina_rb.bic))
        
        #--
        f.write("LOV_PB: {} \n".format(Lov_PB))
        f.write("LOV_DC: {} \n".format(Lov_DC))
        f.close()
        #----------


        previsoes = previsoes.append(simulate_matches_v(forecast,param_PB, param_DC,poisson_model_gol,
                                                        poisson_model_gol_finc,poisson_model_gol_finc_rb,
                                                        poisson_model_gol_fina,poisson_model_gol_fina_rb,
                                                        poisson_model_FinC,logit_model_Prop
                                                        ,scouts, prop,7,20,4),ignore_index=True)
        resultado = previsoes

        
    return resultado        
```

```python
def compara_modelos(acuracias):
    identificador = 2014
    Defi = {'2014':[], '2015':[], '2016':[], '2017':[], '2018':[]}
    Pro = {'2014':[], '2015':[], '2016':[], '2017':[], '2018':[]}
    RPS = {'2014':[], '2015':[], '2016':[], '2017':[], '2018':[]}
    label = {0:'PB_acu', 1:'PD_acu', 2:'DC_acu', 3:'PD_gol_acu',
             4:'PD_gol_FC_acu', 5:'PD_gol_FC_rb_acu', 6:'PD_gol_F_acu', 7:'PD_gol_F_rb_acu', 8:'Fin_acu_pf', 9:'Fin_acu_pl'}
    for result in acuracias:
        aux = []
        aux2 = []
        aux3 = []
        for mod in result:
            aux.append(mod['De Finetti'][19])
            aux2.append(mod['Proporção'][19])
            aux3.append(mod['RPS'][19])
        Finetti = aux.copy()
        Propo = aux2.copy()
        RPS = aux3.copy()
        Finetti.sort()
        RPS.sort()
        Propo.sort(reverse = True)
        for fineti, prop, rps in zip(Finetti, Propo, RPS):
            Defi[str(identificador)].append(label[aux.index(fineti)])
            Pro[str(identificador)].append(label[aux2.index(prop)])
            RPS[str(identificador)].append(label[aux3.index(rps)])
        identificador +=1
    return Defi, Pro, RPS
              
compara_modelos(acuracias)       
```
