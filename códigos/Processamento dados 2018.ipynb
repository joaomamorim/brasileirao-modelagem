```python
import pandas as pd
import numpy as np
```

```python
# lendo a base de cada um dos anos
print('Análise do ano de 2017 \n \n \n')
scouts_2018 = pd.read_csv('D:\\Dropbox\\Projeto Brasileirao\\caRtola\\data\\2018\\2018_scouts_raw.csv')
```

```python
print('Tamanho da base de 2018 antes da limpeza', scouts_2018.shape)
scouts_2018['Ano'] = 2018
# Eliminando jogadores com nan no nome do clube e que não participaram de jogos
scouts_2018 = scouts_2018[(scouts_2018['atletas.variacao_num'] != 0) | (scouts_2018['atletas.pontos_num'] != 0)]
print('Tamanho da base com apenas jogadores que jogaram:',scouts_2018.shape)
scouts_2018_limpo = scouts_2018[(scouts_2018['atletas.rodada_id']>0)]
scouts_2018_limpo = scouts_2018_limpo.drop(['Unnamed: 0', 'Unnamed: 0.1', 'atletas.status_id', 'atletas.nome',
                       'atletas.foto', 'atletas.slug', 'atletas.clube_id'], axis = 1)
scouts_2018_limpo.rename(columns={'atletas.apelido':'Apelido', 'atletas.atleta_id': 'AtletaID'
                                 ,'atletas.jogos_num': 'Jogos'
                                 ,'atletas.media_num':  'PontosMedia',  'atletas.pontos_num':  'Pontos'
                                 ,'atletas.posicao_id':'Posicao', 'atletas.preco_num':'Preco'
                                 ,'atletas.variacao_num':'PrecoVariacao', 'atletas.clube.id.full.name':'Clube',
                                 'atletas.rodada_id': 'Rodada'}, inplace=True)
print('Tamanho da base limpa:', scouts_2018_limpo.shape); print('Considerando que um time usa até 14 jogadores é esperado que em 20 time e 38 tenhamos ', 20*14*38)

scouts_2018_limpo.sample(10)
```

```python
aux = scouts_2018_limpo.loc[:,['AtletaID', 'FS', 'PE', 'A', 'FT', 'FD', 'FF', 'G', 'I',
       'PP', 'RB', 'FC', 'GC', 'CA', 'CV', 'SG', 'DD', 'DP', 'GS']]
```

```python
aux = aux.groupby(by='AtletaID').diff().fillna(aux).replace(0, np.NaN)
scouts_2018_limpo.loc[:,['FS', 'PE', 'A', 'FT', 'FD', 'FF', 'G', 'I',
       'PP', 'RB', 'FC', 'GC', 'CA', 'CV', 'SG', 'DD', 'DP', 'GS']] = aux
```
